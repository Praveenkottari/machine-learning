## Polynomial Regression

- Linear regression fits a straight line to data, but not all data relationships are straight, especially when the data has curves or bends
- To capture these curves, we can create new features by adding higher powers of the original features, like squaring or cubing them ($x^2,x^3,...$)
- These new features allow the model to find patterns that are non-linear in the original data but appear linear when transformed into a higher-dimensional space
- The PolynomialFeatures class in scikit-learn helps create these new polynomial features, making it easy to apply this technique
- After transforming the data, we apply linear regression to the new feature set, allowing the model to learn more complex relationships in a simple way

**It's important to understand that polynomial regression remains a linear model, but when visualized in the original feature space, it appears as a curve. This happens because the hyperplane formed in the higher-dimensional feature space corresponds to a complex curved line in the original space.**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn.linear_model import LinearRegression
```

```python
df= pd.read_csv('poly_dataset.csv')
```

```python
df.isnull().sum()
```

```python
x=df.iloc[:,:-1].values
y=df.iloc[:,-1].values
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(x, y, color="blue", label="samples")
plt.xlabel('X')
plt.ylabel('y')
plt.title('Data samples')
plt.legend()
plt.show()
```

```python
lin_reg=LinearRegression()
lin_reg.fit(x,y)
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly_reg=PolynomialFeatures(degree=2)
x_poly=poly_reg.fit_transform(x)
lin_reg_2=LinearRegression()
lin_reg_2.fit(x_poly,y)
```

```python
print(x_poly)
```

```python
plt.title('input feature vs poly_feature')
plt.plot(x,x_poly[:,0], label = 'order 0')
plt.plot(x,x_poly[:,1], label = 'order 1')
plt.plot(x,x_poly[:,2] , label = 'order 2')

#plt.plot(x,x_poly[:,3] , label = 'order 3')
# plt.plot(x,x_poly[:,4] , label = 'order 4')
plt.xlabel('x')
plt.ylabel('x_poly')
plt.legend()
```

```python
lin_reg_2.predict(poly_reg.fit_transform([[-4]]))
```

```python
plt.scatter(x,y ,color="red")
plt.plot(x,lin_reg.predict(x),color="blue")
plt.title("(linear Regression)")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
plt.scatter(x,y ,color="red")
plt.plot(x,lin_reg_2.predict(poly_reg.fit_transform(x)),color="blue")
plt.title("(Poly Regression)")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```
